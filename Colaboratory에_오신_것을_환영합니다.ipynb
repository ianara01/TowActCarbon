```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
from fipy import CellVariable, Grid1D, TransientTerm, DiffusionTerm, Viewer
from fipy.viewers.matplotlibViewer import Matplotlib1DViewer
import numpy as np
import matplotlib.pyplot as plt

# 문제 설정
nx = 50  # 격자 개수
L = 10.0  # 길이
D = 1.0  # 확산 계수
v = 0.1  # 속도

mesh = Grid1D(nx=nx, dx=L / nx)
C = CellVariable(name="Concentration", mesh=mesh, value=0.0)

# 초기 조건
C.setValue(1.0, where=mesh.cellCenters[0] < L / 5)

# PDE 정의
eq = TransientTerm() + v * C.faceGrad.divergence - DiffusionTerm(coeff=D)

# 시간 루프
time_step = 0.01
steps = 500

viewer = Matplotlib1DViewer(vars=C)
#viewer = Viewer(C, limits=(0, 1))
for step in range(steps):
    eq.solve(var=C, dt=time_step)
viewer.plot()
```

```
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 문제 설정
L = 10.0  # 길이
nx = 50  # 격자 수
dx = L / nx
k = 1.0  # 열전도도
cp = 1.0  # 비열
rho = 1.0  # 밀도
alpha = k / (rho * cp)

# 초기 조건
T0 = np.zeros(nx)
T0[nx // 2] = 100  # 중앙에 온도 초기화

# 열 전달 방정식
def heat_equation(t, T):
    dTdx2 = np.zeros(nx)
    dTdx2[1:-1] = (T[2:] - 2 * T[1:-1] + T[:-2]) / dx**2
    return alpha * dTdx2

# 시간 통합
t_span = (0, 10)
t_eval = np.linspace(0, 10, 500)
sol = solve_ivp(heat_equation, t_span, T0, t_eval=t_eval)

# 결과 시각화
for i in range(0, len(sol.t), 10):
    plt.plot(np.linspace(0, L, nx), sol.y[:, i], label=f"t={sol.t[i]:.2f}")
plt.xlabel("Position")
plt.ylabel("Temperature")
plt.legend()
plt.show()
```

```
from pyomo.environ import ConcreteModel, Var, Objective, Constraint, SolverFactory


model = ConcreteModel()

# 문제 설정
L = 10.0  # 길이
n = 20  # 노드 수
dx = L / (n - 1)
mu = 1.0  # 점성 계수

model.x = range(n)
model.v = Var(model.x, initialize=0.0)
model.P = Var(model.x, initialize=lambda m, i: L - i * dx)

# 운동량 방정식 (Darcy's Law)
def momentum_eqn(m, i):
    if i == 0 or i == n - 1:
        return Constraint.Skip  # 경계 조건
    return (m.v[i+1] - 2 * m.v[i] + m.v[i-1]) / dx**2 == - (m.P[i+1] - m.P[i]) / dx

model.momentum = Constraint(model.x, rule=momentum_eqn)

# 목적 함수 (가상의 문제 정의)
model.obj = Objective(expr=sum((model.v[i] - 1.0)**2 for i in model.x))

solver = SolverFactory('ipopt')
if solver.available():
    print("Solver is available!")
else:
    print("Solver is not available!")
solver.solve(model)

# 결과 출력
v = [model.v[i]() for i in model.x]
P = [model.P[i]() for i in model.x]
```
